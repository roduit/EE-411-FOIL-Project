### Obtain pre computed results stored as pickle files
As the computation of these figures is time expensive, results are stored in pickle format in order to plot without computing each time

```python
# A mettre dans les fonctions de display
def visualize_convergence(losses, test_loss, test_accuracy, num_epochs, width):
    plt.figure(figsize=(12, 6))
    plt.title(f"Training loss for width = {width} and for {num_epochs} Epochs")
    plt.plot(np.linspace(1, num_epochs, num_epochs), losses, label="Training loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.legend()
    plt.show()
```

```python
losses_k4_1K = open_pickle("losses_k_LastVersion.pkl")
test_loss_k4_1K = open_pickle("test_loss_k4_LastVersion.pkl")
test_accuracy_k4_1K = open_pickle("test_accuracy_k4_LastVersion.pkl")

losses_k4_200E = open_pickle("losses_k4_200E.pkl")
test_loss_k4_200E = open_pickle("test_loss_k4_200E.pkl")
test_accuracy_k4_200E = open_pickle("test_accuracy_k4_200E.pkl")
```

```python
visualize_convergence(
    losses_k4_200E, test_loss_k4_200E, test_accuracy_k4_200E, num_epochs=200, width=4
)
```
